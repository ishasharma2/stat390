```python
import pandas as pd
a = pd.read_csv('sarima_predictions.csv')
targets = ['Date','NORTH LAWNDALE','HUMBOLDT PARK','EAST GARFIELD PARK','ENGLEWOOD',
'AUBURN GRESHAM','WEST GARFIELD PARK','ROSELAND','GREATER GRAND CROSSING','WEST PULLMAN',
'WEST ENGLEWOOD','SOUTH SHORE','SOUTH LAWNDALE','CHICAGO LAWN','NEW CITY',
'AUSTIN']
a = a.loc[:, targets]
modified_columns = a.iloc[:, 1:].applymap(lambda x: 0 if x < 0 else x)

# Concatenate the first column with the modified columns
df = pd.concat([a.iloc[:, 0:1], modified_columns], axis=1)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])

sorted_columns = sorted(df.columns[1:])

# Create a new DataFrame with columns in the desired order
df_sorted = df[['Date'] + sorted_columns]

# Filter the DataFrame for 2024 and 2025
df_2024 = df_sorted[df_sorted['Date'].dt.year == 2024]
df_2025 = df_sorted[df_sorted['Date'].dt.year == 2025]

# Calculate the sum of programs for each neighborhood in 2024
sum_2024 = df_2024.iloc[:, 1:].sum()

# Calculate the sum of programs for each neighborhood in 2025
sum_2025 = df_2025.iloc[:, 1:].sum()

# Display the sum of programs for each neighborhood in 2024
print("Program sum for 2024:")
print(sum_2024)

# Display the sum of programs for each neighborhood in 2025
print("Program sum for 2025:")
print(sum_2025)
```
